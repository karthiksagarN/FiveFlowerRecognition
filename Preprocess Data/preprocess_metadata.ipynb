```python
import zipfile
import os

def extract_zip_file(zip_file_path, extract_to_path):
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(extract_to_path)
        print(f"Files extracted successfully to {extract_to_path}")
    except FileNotFoundError:
        print("Zip file not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
zip_file_path = '/content/drive/MyDrive/FiveFlowersRecognition/Dataset/flowers.zip'  # Replace with the actual path to your flowers.zip file
extract_to_path = '/content/drive/MyDrive/FiveFlowersRecognition/Dataset'  # Replace with the path where you want to extract the files

# Create the extraction directory if it doesn't exist
os.makedirs(extract_to_path, exist_ok=True)
extract_zip_file(zip_file_path, extract_to_path)
```

```python
# Function to extract text from a .txt file
def extract_text_from_txt(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return "File not found."
    except Exception as e:
        return str(e)

# Usage example
file_path = '/content/drive/MyDrive/FiveFlowersRecognition/Dataset/labels.txt'
text_content = extract_text_from_txt(file_path)
print(text_content)
```

```python
processed_lines = []
for line in text_content.strip().split('\n'):
    processed_line = line.split('.jpg')[0] + '.jpg'
    processed_lines.append(processed_line)

# Output the processed data
processed_data = '\n'.join(processed_lines)
print(processed_data)
```

```python
# Replace '/' with ','
labeled_data = processed_data.replace('/', ',')

# Print the modified content
print(labeled_data)
```

```python
import os
import pandas as pd

# Path to the labels CSV file
labels_csv_path = '/content/drive/MyDrive/FiveFlowersRecognition/Dataset/flowers/labels.csv'  # Replace with the actual path to your labels.csv file

# Define the paths to your image folders
folders = ['/content/drive/MyDrive/FiveFlowersRecognition/Dataset/flowers/daisy', '/content/drive/MyDrive/FiveFlowersRecognition/Dataset/flowers/dandelion', '/content/drive/MyDrive/FiveFlowersRecognition/Dataset/flowers/rose', '/content/drive/MyDrive/FiveFlowersRecognition/Dataset/flowers/sunflower', '/content/drive/MyDrive/FiveFlowersRecognition/Dataset/flowers/tulip']  # Replace with actual paths

# Read the CSV into a DataFrame
labels_df = pd.read_csv(labels_csv_path)

# Get the list of labeled images
labeled_images = set(labels_df['filename'])

# Iterate through each folder and remove images that are not labeled
for folder in folders:
    for file_name in os.listdir(folder):
        if file_name not in labeled_images:
            file_path = os.path.join(folder, file_name)
            os.remove(file_path)
            print(f"Removed {file_path}")

print("Unlabeled images removed successfully.")
```
